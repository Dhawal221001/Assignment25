## Q2: What is a scatter plot? Use the following code to generate data for x and y. Using this generated data plot a scatter plot.

### A graph in which the values of two variables are plotted along two axes, the pattern of the resulting points revealing any correlation present.

```python
import numpy as np
np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))
```

```python
import matplotlib.pyplot as plt
```

```python
x
```

```python
y
```

```python
plt.scatter(x,y)
plt.xlabel("This is my x-axis")
plt.ylabel("This is my y-axis")
plt.title("this is a scatter plot")
```

## Q3: Why is the subplot() function used? Draw four line plots using the subplot() function.

## subplot() function adds subplot to a current figure at the specified grid position.

```python
#1.
x = np.array([0, 1, 2, 3, 4, 5]) 
y = np.array([0, 100, 200, 300, 400, 500])

plt.subplot(2,2,1)
plt.plot(x,y)

#2.
x = np.array([0, 1, 2, 3, 4, 5]) 
y = np.array([50, 20, 40, 20, 60, 70])

plt.subplot(2,2,2)
plt.plot(x,y)

#3.
x = np.array([0, 1, 2, 3, 4, 5]) 
y = np.array([10, 20, 30, 40, 50, 60])

plt.subplot(2,2,3)
plt.plot(x,y)

#4.
x = np.array([0, 1, 2, 3, 4, 5]) 
y = np.array([200, 350, 250, 550, 450, 150])

plt.subplot(2,2,4)
plt.plot(x,y)


plt.show()
```

## Q4: What is a bar plot? Why is it used? Using the following data plot a bar plot and a horizontal bar plot.

### A bar plot is a plot that presents categorical data with rectangular bars with lengths proportional to the values that they represent.It shows comparisons among discrete categories. 

```python
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])
```

```python
plt.figure(figsize=(3,3))
plt.bar(company,profit)
```

```python
plt.figure(figsize=(3,3))
plt.barh(company,profit)
```

## Q5: What is a box plot? Why is it used? Using the following data plot a box plot.

### A simple way of representing statistical data on a plot in which a rectangle is drawn to represent the second and third quartiles, usually with a horizontal line inside to indicate the median value. The lower and upper quartiles are shown as vertical lines either side of the rectangle.         Box plots are used to show distributions of numeric data values, especially when you want to compare them between multiple groups.

```python
box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)
```

```python
plt.subplot(1,2,1)
plt.boxplot(box1)
plt.title("Box1")

plt.subplot(1,2,2)
plt.boxplot(box2)
plt.title("Box2")

plt.figure(figsize=(6,6))
plt.show()
```
